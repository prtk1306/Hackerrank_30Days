```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
n = int(input())

for _ in range(n):
    
    s = input()
    
    o = ''
    e = ''
    
    for i in range(0,len(s),2):

        o += s[i]
        try:
            e += s[i+1]
        except:
            pass
    print(o + ' ' + e)
```
